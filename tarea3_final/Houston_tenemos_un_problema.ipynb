```python
# using the ExcelFile class
#Lee el archivo con pandas y lo guarda las columnas importantes en murders
xls = pd.ExcelFile('Houston-security.xls')
murders=xls.parse('rptPublicAffairs', parse_cols= [1,2], na_values=['NA'])

```

```python
# se ingresan df ( que es murders normalmente) y un string s.
#retorna 2 listas. d las horas y deads el numero de muetes por cada hora para el string s
def darMuertes(df, s):
    horas = arange(24)
    deads = zeros(24)
    d= []
    for x in range(0,num_rows-1):
        offense = df.iloc[x,1]
        if(offense == s):
            hour = df.iloc[x,0]
            h = int(hour)
            deads[h] = deads[h] +1
            d.append(h)
    return [d,deads]
```

```python
# lee la lista deads y da la hora mas segura 
deads = darMuertes(murders,'Aggravated Assault')[1]
m = 0
for x in range(0,24):
    if(deads[x] == deads.min()):
        m = x
        break
print "La hora más segura para salir según asalto agravado es a las %d: " % m

```

```python
#Hace el histograma segun el string s
def hacerHistograma(murders,s):
    d = darMuertes(murders,s)[0]
    deads = darMuertes(murders,s)[1]
    hist(d, bins=24, facecolor='green', normed=False, color='b', alpha=1)    
    xlabel('Horas')
    ylabel('Asesinatos')
    title(s)
    axis([0, 23, 0, max(deads)+5])
    grid(False)
    show()
```
